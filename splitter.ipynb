```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
```

```python
model = SentenceTransformer("all-mpnet-base-v2")
```

```python
final = pd.read_csv("real_final.csv", index_col = 0)
```

```python
final.head()
```

```python
final = final.fillna('')
```

```python
final.head()
```

```python
execs = final["Execution Modifiers"].to_list()
style = final["Stylistic Modifiers"].to_list()
content = final["Core Content"].to_list()
```

```python
content_embs = []
for item in tqdm(content):
    if len(content) == 0:
        content_embs.append({"Core Content": item, "content_embedding": ''})
    else:
        embedding = model.encode(item)
        content_embs.append({"Core Content": item, "content_embedding": embedding})
```

```python
style_embs = []
for item in tqdm(style):
    if len(style) == 0:
        style_embs.append({"Stylistic Modifiers": item, "style_embedding": ''})
    else:
        embedding = model.encode(item)
        style_embs.append({"Stylistic Modifiers": item, "style_embedding": embedding})
```

```python
execs_embs = []
for item in tqdm(execs):
    if len(execs) == 0:
        execs_embs.append({"Execution Modifiers": item, "exec_embedding": ''})
    else:
        embedding = model.encode(item)
        execs_embs.append({"Execution Modifiers": item, "exec_embedding": embedding})
```

```python
content_embs[0]
```

```python
content_df = pd.DataFrame(content_embs)
style_df = pd.DataFrame(style_embs)
exec_df = pd.DataFrame(execs_embs)
```
